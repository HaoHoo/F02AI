通常来说，完成一个机器学习任务，首先需要有数据。在我们这个样例里面，需要使用做过标记的数据（这种使用进行标记的数据对AI进行训练的方式，也叫监督学习）。对于很大一部分机器学习，会把提供的数据分为两个部分，一部分用来训练人工智能模型，一部分用来验证训练的结果。
因此，在上图中我们就可以看到，机器学习的常见流程中，首先需要导入数据。如果数据不是现成的数据集，就会对数据进行清洗和处理，去除不需要的、缺失的以及会影响学习结果的数据。然后，会将数据按照比例（例如80:20）拆分为两部分，多的部分用于训练，少的部分用于验证。
准备好数据后，按照不同的机器学习需求，我们需要选择合适的算法。然后，将算法和训练数据关联起来，形成一个训练模型。而机器学习的过程，就是按照这个模型去训练计算机，并且使用评分模型来评价这个模型的准确性。

这只是机器学习工作室最简单的一个例子，更为完整的介绍可以参考：[Microsoft Azure 机器学习工作室功能概述](https://download.microsoft.com/download/C/4/6/C4606116-522F-428A-BE04-B6D3213E9E52/ml_studio_overview_v1.1.pdf)

实际上，按照需要解决问题类型的不同，机器学习会选择不同的算法。下面就简单介绍几种常见的算法。

**分类算法**（Classification Algorithms）
分类算法常用于判断，可以有两个或多个分类，解决“这是A还是B？”或者“这是A、B、C还是D？“这类的问题。比如，油箱里的油在这条路上还能跑到加油站吗？或者，哪一种优惠措施会带来更多的营收？

**异常检测**（Anomaly Detection Algorithms）
异常检测常用于判断是否出现异常，例如信用卡公司会按照你多维度的信息，判断你是否正常使用信用卡，乃至提醒你“奇怪的”刷卡行为，是否存在欺诈等。利用人工智能分析多维度数据，能够帮助我们在很多领域解决异常发现的问题，比如收集飞机发动机的很多数据，在出现故障之前就提醒人们进行处理。

**回归算法**（Regression Algorithms）
回归算法是一种重要的预测算法，可以通过数据的机器学习，帮助我们预测数值。例如，通过对天气数据的学习，预测接下来的天气、气温等。能够帮助我们解答关于“多少”的数字问题。
         
**聚类算法**（Clustering Algorithms）
和其他算法有点不一样，聚类算法是“无监督学习”的算法。即，数据本身并没有进行标记，而是机器学习在不断的训练中，自行“发现”数据的规律，对数据进行分类。例如，聚类可以帮助我们向用户推荐相近的产品。

**强化学习**（Reinforcement Learning Algorithms）
强化学习来自于对大脑活动的模仿，不同的行为对应了惩罚或奖励，从而让算法从中学习，来决定后续活动。强化学习适合自动化系统，在无人干预的情况下作出大量小决策。

当然，实际使用的算法远远不止这些，还有更加细化的分类。例如微软的机器学习工作室，就将算法分了五个大类，每一类下面再做细分。

这里就是最简单的MNIST的机器学习模型在Azure机器学习工作室中的样子。我们来仔细看看。
首先，为了便于试验，Azure机器学习工作室已经把MNIST的数据集做了初步的处理。处理后的数据集，其实就和上一个记事本中看的数据集一样。您可以在机器学习工作室中，找到“Saved Dataset”，“Samples”，然后找到MNIST的数据集，直接拖放到画布上。需要注意的是，数据集已经帮您切分为6万条训练数据和1万条验证数据。依次添加两个数据集。您可以在数据集模块上右键，选择“Dataset”，“Visualize”，查看数据的展现形式，和上一个记事本比较。

接下来，我们需要拖放一个机器学习算法模块。找到“Machine Learning”，然后找到”Initialize Model“，找到“Classification“，将其中的“MultiClass Neural Network“拖放到画布中。这样，就添加了机器学习算法模块。

然后，我们要放置一个训练模型和一个评分模型。查找的路径为“Machine Learning”，“Train”，“Train Model”；和“Machine Learning”，“Score”，“Score Model”。

现在，我们的机器学习模型已具雏形。为了评估机器学习的准确性，我们可以继续添加一个评估模型和一个脚本模块。评估模型能够对机器学习的结果进行评估，而脚本模块能将评估结果最终显示为一个评分。其查找路径分别为“Machine Learning“，”Evalute“，”Evalute Model“；和”R Lanuage Modules“，”Execute R Script“。

按照上图，将各个模块连接起来。接下来我们需要对各个模块进行一些配置。

首先我们选中“MultiClass Neural Network”，在右边的属性框会显示所有可配置的参数。“Create trainer mode“我们选择”Single Parameter“，”Hidden layer specification“我们选择”Custom difinition script“。这表明这个是一个单一单数的隐含层，使用定制化的脚本来描述隐含层模型。
接下来我们在“Neural network definition”框内输入定制化的脚本。

```
input Picture [28,28];
hidden H [100] from Picture all;
output Result [10] softmax from H all;
```

这表明该隐含层的输入是一个28x28的矩阵，正好是MNIST中对手写数字图片的定义。而隐含层 H 则是基于输入的100维神经网络层，输出最终是需要判断具体数字，因此使用了 softmax 函数针对 H 隐含层输出10个类别。之所以选择“MultiClass Neural Network”这个算法模块，也正是基于我们需要获得 0-9 数字判断的需求。

接下来配置一些其他参数：

参数|值
-|:-:|-:
The learning rate|0.1
Number of learning iterations|30
The initial learning weights diameter|0.1
The momentum|0
The type of normalizer|Min-Max normalizer
Random number seed|1

接下来我们配置训练模块的参数。训练模块参数很简单，我们需要告诉机器学习工作室，哪一列数据用于训练输入的数据。很显然，我们的数据集是一个784(28x28)+1列的表格。784列中是每一个手写数字图片的像素，而增加的一列 Label，就是这个手写数字对应的 0-9 的数字。所以，我们在训练模块中，选择这一列。

评分模块只需要钩上“Append score columns to output”即可。这会附加评分的列到表格，便于后面我们评估。

评估模型不需要额外的配置。

最后我们需要给“Execute R Script“模块输入R语言的脚本：

```
dataset1 <- maml.mapInputPort(1)
feat <- data.matrix(dataset1[,2:11])
total = sum(sum(feat))
correct = sum(diag(feat))
accuracy = correct / total
acc = as.data.frame(accuracy)
maml.mapOutputPort("acc");
```

连接好所有的模块，完成了全部的参数配置，我们可以运行这个机器学习模型了。在最下方的界面，点击“Run”，开始运行。
如果一切正常，每一个模块完成运行后，都会打上一个勾。如果出现错误，请仔细检查模块选择是否正确，是否正确的配置了参数。

耐性等待，最后我们能够看到得到的机器学习模型的准确度。
